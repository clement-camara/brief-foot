```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
URL = 'https://www.lequipe.fr/Football/ligue-1/page-participants'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find_all('ul', class_='clubs')

list_url_clubs = []
for result in results:
    liens_clubs = result.find_all('ul', class_='club-links')
    for liens in liens_clubs:
        list_url_clubs.append(liens)
        
list_liens = []
for club in list_url_clubs:
    lien = club.find('a', class_='link')
    list_liens.append(lien.get('href'))
print(list_liens)
```

```python
rows_players = []
for i in range(len(list_dfs)):
    try:
        for j in range(len(list_dfs[i][2]['Nom'])):
            list_dfs[i][2].set_axis(
                ['num', 'Nom', 'nationality', 'birthdate', 'position'], axis=1, inplace=True)
            dict_players = {}
            dict_players['full_name'] = list_dfs[i][2]['Nom'][j]
            dict_players['birthdate'] = list_dfs[i][2]['birthdate'][j]
            dict_players['team_id'] = list_dfs[i][0][0][0].split(':')[1].strip()
            dict_players['position'] = list_dfs[i][2]['position'][j]
            dict_players['nationality'] = list_dfs[i][2]['nationality'][j]
            rows_players.append(dict_players)
    except:
        for j in range(len(list_dfs[i][1]['Nom'])):
            list_dfs[i][1].set_axis(
              ['num', 'Nom', 'nationality', 'birthdate', 'position'], axis=1, inplace=True)
            dict_players = {}
            dict_players['full_name'] = list_dfs[i][1]['Nom'][j]
            dict_players['birthdate'] = list_dfs[i][1]['birthdate'][j]
            dict_players['team_id'] = list_dfs[i][0][0][0].split(':')[1].strip()
            dict_players['position'] = list_dfs[i][1]['position'][j]
            dict_players['nationality'] = list_dfs[i][1]['nationality'][j]
            rows_players.append(dict_players)
rows_players[0]
```

```python
df_players = pd.DataFrame(rows_players)
df_players
```
